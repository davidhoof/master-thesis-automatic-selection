```python
import multiprocessing
import time
import glob
import os
import shutil
```

```python
num_workers=multiprocessing.cpu_count()*2
load_root="FractalDB-Pretrained-ResNet-PyTorch/data/csv_rate0.2_category60/"
save_root="FractalDB-Pretrained-ResNet-PyTorch/data/FractalDB"
temp_root="FractalDB-Pretrained-ResNet-PyTorch/data/csv_rate0.2_category60_tmp/"
```

```python
def split(a, n):
    k, m = divmod(len(a), n)
    return (a[i*k+min(i, m):(i+1)*k+min(i+1, m)] for i in range(n))
```

```python
path_list = glob.glob(os.path.join(load_root,"*.csv"))
path_list_parted = list(split(path_list,num_workers))
```

```python
worker=0
work_paths=[]
for paths in path_list_parted:
    if not os.path.exists(temp_root):
        os.mkdir(temp_root)
    dst_dir=os.path.join(temp_root,str(worker))
    os.mkdir(dst_dir)
    for path in paths:
        shutil.copyfile(path, os.path.join(dst_dir,os.path.basename(path)))
    work_paths.append(os.path.join(temp_root,str(worker)))  
    worker=worker+1
```

```python
work_paths
```

```python
def mp_worker(path):
    !python FractalDB-Pretrained-ResNet-PyTorch/fractal_renderer/make_fractaldb.py --load_root $path --save_root $save_root --weight_csv "FractalDB-Pretrained-ResNet-PyTorch/fractal_renderer/weights/weights_0.1.csv" --image_size_x 32 --image_size_y 32 --pad_size_x 2 --pad_size_y 2 --instance 10 --draw_type "patch_color"

def mp_handler():
    p = multiprocessing.Pool(num_workers)
    p.map(mp_worker, work_paths)

mp_handler()
```
