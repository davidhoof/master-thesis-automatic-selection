```python
def get_children(model: torch.nn.Module):
    # get children form model!
    children = list(model.children())
    flatt_children = []
    if children == []:
        # if model has no children; model is last child! :O
        return model
    else:
       # look for children from children... to the last child!
       for child in children:
            try:
                flatt_children.extend(get_children(child))
            except TypeError:
                flatt_children.append(get_children(child))
    return flatt_children
```

```python
class LayerHook():
    def __init__(self):
        self.storage = None
        self.hook_handle = None

    def pull(self):
        if self.hook_handle is not None:
            self.hook_handle.remove()
        return self.storage

    def register_hook(self, module, store_input=True):
        if self.hook_handle is not None:
            self.hook_handle.remove()
        self.storage = None
        def hook(model, inp, out):
            if store_input:
                self.storage = inp
            else:
                self.storage = out
        self.hook_handle = module.register_forward_hook(hook)
```

```python
pretrained_output_path="output/"
finetuned_output_path="output_sweep_complete/"
models=["lowres_resnet9","lowres_resnet50","lowres_densenet121","lowres_vgg16_bn"]
datasets=["cifar10","cifar100","svhn","tinyimagenet"]
checkpoints=[
    (0,"cifar10_model"),
    (1,"cifar100_model"),
    # (2,"grocerystore_model"),
    (3,"svhn_model"),
    (4,"tinyimagenet_model")
]
compare_checkpoint=5

dataset_path="../SSD/datasets"
output_path="../SSD/output"


def calculate_weight_quality_metrics(model, sparsity_eps=0.01):
    
    info_dict = measure_quality(model,sparsity_eps=sparsity_eps)    
    
    sparsity=[]
    variance_entropy=[]
    variance_entropy_clean=[]
    weights=[]
    
    for key_l,layer in info_dict.items():
        sparsity.append(layer['sparsity'])
        variance_entropy.append(layer['variance_entropy'])
        variance_entropy_clean.append(layer['variance_entropy_clean'])
        weights.append(layer['n'])        
    
    return {
        "sparsity" : np.average(sparsity),
        "variance_entropy" : np.average(variance_entropy),
        "variance_entropy_clean": np.average(variance_entropy_clean),
        "weighted_sparsity" : np.average(a=sparsity, weights=weights),
        "weighted_variance_entropy" : np.average(a=variance_entropy, weights=weights),      
        "weighted_variance_entropy_clean" : np.average(a=variance_entropy_clean, weights=weights), 
           }

def get_silhouette_score_from_latent_space(model):
    
    # load data
    
    data_dir = os.path.join(model.myhparams["data_dir"], model.myhparams["dataset"])
    data_ = None

    if model.myhparams['dataset_percentage'] == 100:
        data_ = data.get_dataset(model.myhparams["dataset"])(data_dir, model.myhparams["batch_size"], model.myhparams["num_workers"])
    if 100 > model.myhparams['dataset_percentage'] > 0:
        data_ = data.get_dataset_minimized(model.myhparams["dataset"])(data_dir, model.myhparams["batch_size"], model.myhparams["num_workers"],
                                                               model.myhparams['dataset_percentage'])
    
    # register forward hook on latent space (bevor classifier)
    hook = LayerHook();
    hook.register_hook(model.model.fc)
    
    # train a batch
    model(next(iter(data_.train_dataloader())))    
    
    latent_space = hook.pull()[0].detach()
        
    # if len(np.shape(latent_space)) == 4:
    print(model.myhparams['classifier'])
    print(np.shape(latent_space))
    
    # if len(np.shape(latent_space)) == 4:
    #     latent_space=latent_space.reshape(np.shape(latent_space)[0] * np.shape(latent_space)[2], np.shape(latent_space)[1] * np.shape(latent_space)[3])    
    
    # calculate silhouette score with kmeans
    km = KMeans(n_clusters=model.myhparams['num_classes'], random_state=42)
    
    km.fit_predict(latent_space)
    
    score = silhouette_score(latent_space, km.labels_, metric='cosine')
    
    return score

def calculate_architecture_size(model):
    children = get_children(model)
    
    names = list(dict.fromkeys([item.__class__.__name__ for item in children]))
    
    architecture_size_dict = {f"architecture_size_{name}": len([item for item in children if str(item.__class__.__name__) == name]) for name in names}
    architecture_size_dict['architecture_size']=len(children)
    
    return architecture_size_dict

def get_metrics(path,model,dataset,checkpoint):
    with open(os.path.join(path,dataset,model,f"version_{checkpoint[0]}","hparams.yaml"), 'r') as stream:
        data_loaded = yaml.safe_load(stream)

    hparams=data_loaded['hparams']
    
    hparams.update({"data_dir":dataset_path})
    hparams.update({"output_dir":output_path})
    
    if not "start_step" in hparams:
        hparams['start_step']=0
    if not "dataset_percentage" in hparams:
        hparams['dataset_percentage']=100
    
    
    # print(hparams)

    loaded_model = TrainModule(hparams)
    # print(glob.glob(os.path.join(path,dataset,model,f"version_{checkpoint[0]}","checkpoints","epoch*.ckpt"))[0])
    state = torch.load(glob.glob(os.path.join(path,dataset,model,f"version_{checkpoint[0]}","checkpoints","epoch*.ckpt"))[0], map_location=loaded_model.device)       
    
    if "state_dict" in state:
        state = state["state_dict"]
        
    loaded_model.model.load_state_dict(
        dict((key.replace("model.classifier.", "model.fc.").replace("model.", ""), value) for (key, value) in
             state.items()))   
    
    silhouette_score = get_silhouette_score_from_latent_space(loaded_model)    
    
    quality_metric_dict = calculate_weight_quality_metrics(loaded_model,sparsity_eps=0.1)
    
    architecture_size_dict = calculate_architecture_size(loaded_model)
    
    # return
    
    metrics = {}
    metrics.update(quality_metric_dict)
    metrics.update(architecture_size_dict)
    metrics['silhouette_score']=silhouette_score
    metrics['pretrained_dataset']=hparams['dataset']      
    
    # print(metrics)
    
    return metrics

def get_difference(path,model,dataset,checkpoint):
    df_tmp = pd.read_csv(os.path.join(path,dataset,model,f"version_{checkpoint[0]}","metrics.csv"))
    acc=float(df_tmp[df_tmp['acc/val'] == df_tmp['acc/val'].max()]['acc/val'].values[0])
    
    df_compare_tmp = pd.read_csv(os.path.join(path,dataset,model,f"version_{compare_checkpoint}","metrics.csv"))
    acc_compare=float(df_compare_tmp[df_compare_tmp['acc/val'] == df_compare_tmp['acc/val'].max()]['acc/val'].values[0])
    return acc-acc_compare    
```

```python
records=[]
for dataset in datasets:
    for model in models:
        for checkpoint in checkpoints:
            difference=get_difference(finetuned_output_path,model,dataset,checkpoint)
            metrics=get_metrics(pretrained_output_path,model,dataset,(0,"Standard"))
            
            # sum the total amout of pooling layer
            if "architecture_size_MaxPool2d" in metrics:
                architecture_size_pool=metrics['architecture_size_MaxPool2d']
            if "architecture_size_AveragePool2d" in metrics:
                architecture_size_pool=metrics['architecture_size_MaxPool2d']
            if "architecture_size_MaxPool2d" in metrics and "architecture_size_AveragePool2d" in metrics:
                architecture_size_pool=metrics['architecture_size_MaxPool2d']+metrics['architecture_size_MaxPool2d']
            
            records.append({
                "checkpoint":checkpoint[1],
                "finetune_dataset":dataset,
                "model":model,
                "difference": difference,
                "pretrained_dataset": metrics['pretrained_dataset'],
                "sparsity": metrics['sparsity'],
                "variance_entropy": metrics['variance_entropy'],
                "variance_entropy_clean": metrics['variance_entropy_clean'],
                "weighted_sparsity": metrics['weighted_sparsity'],
                "weighted_variance_entropy": metrics['weighted_variance_entropy'],
                "weighted_variance_entropy_clean": metrics['weighted_variance_entropy_clean'],
                "architecture_size": metrics['architecture_size'],
                "architecture_size_conv": metrics['architecture_size_Conv2d'],
                "architecture_size_pool": architecture_size_pool,
                "architecture_size_relu": metrics['architecture_size_ReLU'],
                "silhouette_score": metrics["silhouette_score"]
                
            })
df=pd.DataFrame.from_records(records)
```

```python
class RandomForestPipeline:
    def __init__(self):
        # Number of trees in random forest
        n_estimators = [int(x) for x in np.linspace(start = 2, stop = 200, num = 30)]
        # Number of features to consider at every split
        max_features = ['auto', 'sqrt']
        # Maximum number of levels in tree
        max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
        max_depth.append(None)
        # Minimum number of samples required to split a node
        min_samples_split = [2, 5, 10]
        # Minimum number of samples required at each leaf node
        min_samples_leaf = [1, 2, 4]
        # Method of selecting samples for training each tree
        bootstrap = [True, False]
        # Create the random grid
        self.random_grid = {'rf__n_estimators': n_estimators,
                           # 'rf__max_features': max_features,
                           'rf__max_depth': max_depth,
                           'rf__min_samples_split': min_samples_split,
                           'rf__min_samples_leaf': min_samples_leaf,
                           'rf__bootstrap': bootstrap}
        
        self.rf = Pipeline([
            ("sc",StandardScaler()),
            ("rf",RandomForestRegressor(max_depth=4, random_state=0))
        ])
        
        self.regr=None
        
    def fit(self,X,y):
        # Random search of parameters, using 3 fold cross validation, 
        # search across 100 different combinations, and use all available cores
        self.rf_random = RandomizedSearchCV(estimator = self.rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
        # Fit the random search model
        self.rf_random.fit(X, y)
        
        self.regr = self.rf_random.best_estimator_
        
    def plot_importances(self, X, y):
        if self.regr == None:
            raise ValueError("Model must be fitted bevor usage")
        
        importances = self.regr['rf'].feature_importances_
        std = np.std([tree.feature_importances_ for tree in self.regr['rf'].estimators_], axis=0)
        
        forest_importances = pd.Series(importances, index=features).sort_values(ascending=False)

        fig, ax = plt.subplots(1,2,figsize=(20,7))
        forest_importances.plot.bar(yerr=std, ax=ax[0])
        ax[0].set_title("Feature importances using MDI")
        ax[0].set_ylabel("Mean decrease in impurity")
        
        result = permutation_importance(
            self.regr, X, y, n_repeats=10, random_state=42, n_jobs=-1
        )
        forest_importances = pd.Series(result.importances_mean, index=features).sort_values(ascending=False)
        
        forest_importances.plot.bar(yerr=result.importances_std, ax=ax[1])
        ax[1].set_title("Feature importances using permutation on full model")
        ax[1].set_ylabel("Mean accuracy decrease")
        # fig.tight_layout()
        plt.show()
        
    def score(self,X,y):
        return self.regr.score(X,y)
        
```
