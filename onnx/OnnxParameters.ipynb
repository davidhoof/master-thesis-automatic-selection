```python
import onnx
import sys
import os
from tqdm import tqdm
from onnx_opcounter import calculate_params
import onnx
```

```python
import glob
paths = glob.glob('/content/models/onnx_zoo/**/*.onnx', recursive=True)
```

```python
len(paths)
```

```python
file_dict={os.path.basename(path)[:-5] : str(int(calculate_params(onnx.load_model(path)))) for path in tqdm(paths)}
```

```python
import pandas as pd
df=pd.DataFrame.from_dict(file_dict, orient='index')
df.reset_index(inplace=True)
df.rename(columns={"index":"Name",0:"Parameters"},inplace=True)
```

```python
df
```

```python
df.to_csv("parameters_onnx_zoo.csv")
```
